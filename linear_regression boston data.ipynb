# sklearn


```python
import sklearn
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
# train_test_split = it split the data in around 75:25 approx ratio in training and testing set
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,boston_df.Price,test_size=0.2)
#test_size=0.2 by default it distribute 379:127 but now 80%:20%
```

```python
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

```python
lreg=LinearRegression()
```

```python
lreg.fit(X_train,Y_train)
```

```python
pred_train=lreg.predict(X_train)
```

```python
pred_test=lreg.predict(X_test)
```

```python
print(Y_train[0:11],pred_train[0:11])
```

```python
print(X_train[0:11])
```

```python
print('Fit a model X_train and calculate the RMSE with y_train: %.2f'%np.sqrt(np.mean((Y_train-pred_train)**2)))
#.2f is for 2 digit floating point no.
```

```python
print('Fit a model X_test and calculate the RMSE with Y_test: %.2f'%np.sqrt(np.mean((Y_test-pred_test)**2)))
```

```python
test=plt.scatter(pred_test,(pred_test-Y_test),c='r',alpha=0.9)
train=plt.scatter(pred_train,(pred_train-Y_train),c='b',alpha=0.5)
plt.hlines(y=0,xmin=-10,xmax=50)       #hlines is for horizontal lines

plt.legend((train,test),('Training','Test'),loc='lower right')

plt.title('residual plots')
```
