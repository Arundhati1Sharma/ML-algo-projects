```python
import seaborn as sns
data1=sns.load_dataset('tips')  ##tips is a dataset
data1.head()
```

```python
sns.distplot(data1['total_bill'])            ##dist is combination of histogram and normal graph
                                            #c urve depicts the distance of point from the mean value
                                            # kde=False in the argument ... removes the curve from the graph depiction
```

```python
sns.jointplot(x='total_bill',y='tip',data=data1)     ##it is the combination of bar and scatter
```

```python
sns.jointplot(x='total_bill',y='tip',data=data1,kind='reg')        ##kind here gives the graph of linear regression
```

```python
sns.violinplot(x='sex',y='total_bill',hue='time',data=data1,split=True)       ##hue selects a unique factor
```

```python
sns.violinplot(x='sex',y='total_bill',hue='day',data=data1)
```

```python
sns.pairplot(data1,hue='smoker')
```

```python
sns.swarmplot(x='sex',y='total_bill',hue='time',data=data1)
sns.violinplot(x='sex',y='total_bill',hue='time',data=data1,split=True,color='blue')
```
