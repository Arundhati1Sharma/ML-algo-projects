```python
#gradient descent
import pandas as pd 
import matplotlib.pyplot as plt
df=pd.read_csv('C:\\Users\\Vaibhav\\Downloads\\HR_comma_sep.csv')
df.head()
```

```python
df1=df.drop(['last_evaluation' ,'number_project' ,'average_montly_hours','time_spend_company', 'Work_accident','Department'],axis=1)
df1.head()
```

```python
x=df1.drop('left',axis=1)
x
```

```python
y=df1.left
y
```

```python
sal_dum=pd.get_dummies(df1.salary,prefix='salary')
sal_dum             ##dummies r made so that ...ml do not get confused in the data ...
                    #since if taken low_sal=1 and med_sal=2 and high_sal=3 ... ml might think high_sal=mid_sal+low_sal
                                                                                              #since 3=2+1
```

```python
x=df1.drop('salary',axis=1)
x=pd.concat([x,sal_dum],axis=1)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=10)
print(x_train,"\n\n\n",x_test,"\n\n\n",y_train,"\n\n\n",y_test)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_pre=model.predict(x_test)
y_pre
```
