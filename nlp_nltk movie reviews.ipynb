```python
romeo_text="""in other words , don't dismiss this film because of its source . 
if you can get past the whole comic book thing , you might find another stumbling block in from hell's directors , albert and allen hughes . 
getting the hughes brothers to direct this seems almost as ludicrous as casting carrot top in , well , anything , but riddle me this : who better to direct a film that's set in the ghetto and features really violent street crime than the mad geniuses behind menace ii society ?""" 
```

```python
romeo_text.split()
```

```python
nltk.download("punkt")
```

```python
romeo_words=nltk.word_tokenize(romeo_text)       #tokenize removes white spaces ... and we can't use here strip 
                                                   #as it is only used in end or at beginning
romeo_words
```

```python
movie_reviews.words(fileids=positive_fields[0])
```

```python
print(romeo_words)
{word:True for word in romeo_words}          #true word is used since dict is can't b made directly 
                                      #it must have had some value ..it could b anything we could also put our names either
```

```python
def build_bag_of_words_features(words):
    return {word:True for word in words}
build_bag_of_words_features(romeo_words)
```

```python
useless_words=nltk.corpus.stopwords.words("english")           ######english is dataset here...i.e. why got passed
useless_words
```

```python
type(useless_words)
```

```python
def build_bag_of_words_features_filtered(words):
    return {
        word:1 for word in words \
        if not word in useless_words}
build_bag_of_words_features_filtered(romeo_words)
```

```python
all_words=movie_reviews.words()
len(all_words)/1e6
```

```python
filtered_words=[word for word in movie_reviews.words() if not word in useless_words]
(filtered_words)
```

```python
len(filtered_words)/1e6
```

```python
from collections import Counter
word_counter=Counter(filtered_words)
```

```python
most_common_words=word_counter.most_common()[:10]
most_common_words
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
sorted_word_counts=sorted(list(word_counter.values()),reverse=True)

plt.loglog(sorted_word_counts)
plt.ylabel("Freq")
plt.xlabel("Word Rank");
```

```python
plt.hist(sorted_word_counts,bins=50);     #*********************************;*********************#
```

```python
negative_features=[(build_bag_of_words_features_filtered(movie_reviews.words(fileids=[f])),'neg') for f in negative_fields]
print(negative_features[3])
```

```python
positive_features=[(build_bag_of_words_features_filtered(movie_reviews.words(fileids=[f])),'pos') for f in positive_fields]
print(positive_features[6])
```

```python
from nltk.classify import NaiveBayesClassifier
```

```python
split=800
```

```python
sentiment_classifier=NaiveBayesClassifier.train(positive_features[:split]+negative_features[:split])
```

```python
nltk.classify.util.accuracy(sentiment_classifier,positive_features[:split]+negative_features[:split])*100
```

```python
nltk.classify.util.accuracy(sentiment_classifier,positive_features[split:]+negative_features[split:])*100
```
