```python
##classification 
#decision_tree
import pandas as pd 
data=pd.read_csv('C:\\Users\\Vaibhav\\Downloads\\salaries.csv')
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
company_n=LabelEncoder()            ##label encoder converts string into numeric value
job_n=LabelEncoder()
degree_n=LabelEncoder()

data['company_n']=company_n.fit_transform(data.company)
data['job_n']=company_n.fit_transform(data.job)
data['degree_n']=company_n.fit_transform(data.degree)
data
```

```python
df=data.drop(['company','job','degree'],axis=1)
df.head()
```

```python
x=df.drop('salary_more_then_100k',axis=1)
y=df['salary_more_then_100k']
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
print(x_train,"\n\n\n",x_test,"\n\n\n",y_train,"\n\n\n",y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
x_test
```

```python
model.score(x_test,y_test)
```
