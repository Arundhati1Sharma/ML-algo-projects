```python
##slicing in arrays 
import numpy as np
l=np.array([1,2,3,4])
l1=np.array([3,4,5,6])
l2=np.array([4,5,6,7])
a=np.array([l,l1,l2])
l1[1:2:1]              ##slicing is same as *list* in 1d array
                       ##for slicing in 2d array ... keep in mind that elements internally are stored in 1d form 
                       ##1st element in slicing of 2d array denotes the row value
print(a[0,0])           ## 1 would b d output 
print(a([1:3],[1:3]))
print(a([0:2],[2,4]))
print(a([0,0],[0,2],[1,0],[1,2],[2,0],[2,2]))
print(a([0:3],[0:3:+2]))
```
