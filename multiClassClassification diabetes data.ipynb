```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.datasets import load_diabetes
```

```python
diabetes=load_diabetes()
print(diabetes.DESCR)
```

load_diabetes has 'data','feature_names','DESCR','target' etc.

```python
diabetes_df=DataFrame(diabetes.data)
diabetes_df.columns=diabetes.feature_names
diabetes_df
```

```python
diabetes_df['Diabetic_level']=diabetes.target
diabetes_df[:10]
```

```python
def diabetic_check(x):
    if x>126:
        return 1
    else:
        return 0
```

```python
diabetes_df['Diabetes']=diabetes_df['Diabetic_level'].apply(diabetic_check)
diabetes_df[:10]
```

```python
plt.figure(figsize=(20,10))
sns.countplot('bp',data=diabetes_df,hue='Diabetes',palette='coolwarm')
```

```python
X=diabetes_df.drop(['age','sex','s1','s2','s3','s4','s5','s6','Diabetic_level','Diabetes'],axis=1)
X[:10]
```

```python
X.shape
```

```python
Y=diabetes_df.Diabetes
Y.head()
```

# sklearn.linear_model              sklearn.model_selection

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
logreg=LogisticRegression()
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.4,random_state=2) #random state
logreg.fit(X_train,Y_train)
Y_pred=logreg.predict(X_test)
```

```python
from sklearn import metrics
print(metrics.accuracy_score(Y_test,Y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier

list_accuracy=[]
for i in range (1,40):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,Y_train)
    Y_pred_knn=knn.predict(X_test)
    list_accuracy.append(metrics.accuracy_score(Y_test,Y_pred_knn))
    
```

```python
plt.plot(range(1,40),list_accuracy)
```

```python
knn=KNeighborsClassifier(n_neighbors=34)
knn.fit(X_train,Y_train)
Y_pred_knn=knn.predict(X_test)
print(metrics.accuracy_score(Y_test,Y_pred_knn))
```
