```python
total_survived_person=data[data['Survived']==1]['Sex'].value_counts()
print(total_survived_person)
```

```python
death_person=data[data['Survived']==0]['Sex'].value_counts()
print(death_person)
```

```python
data1=pd.DataFrame([total_survived_person,death_person])
data1.index=['Survived','death']
data1
```

```python
import matplotlib.pyplot as plt 
data1.plot(kind='bar',title='bar plot')
plt.show()
```
