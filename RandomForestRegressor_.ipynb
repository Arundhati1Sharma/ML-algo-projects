```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
```

```python
x=10*np.random.rand(100)
x
```

```python
x.shape
```

```python
def sin_model(x,sigma=0.2):
    noise=sigma*np.random.randn(len(x))
    return np.sin(5*x)+np.sin(0.5*x)+noise
```

```python
y=sin_model(x)
plt.figure(figsize=(16,8))
plt.errorbar(x,y,0.1,fmt='s')
```

```python
xfit=np.linspace(0,10,1000)
rfr=RandomForestRegressor(150)
rfr.fit(x[: ,None],y)
yfit=rfr.predict(xfit[:,None])
ytrue=sin_model(xfit,0)
plt.figure(figsize=(16,8))
plt.errorbar(x,y,0.1,fmt='o')   # 0.1 is the size of bar/line on points and fmt denotes format here circle for o, square for s
plt.plot(xfit,yfit,'-r',alpha=0.8)
plt.plot(xfit,ytrue,'-b',alpha=0.9)  #alpha represent darkness
```
