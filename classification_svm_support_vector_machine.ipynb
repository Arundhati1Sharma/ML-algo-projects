```python
x=dframe.drop('target',axis=1)
x.head()
```

```python
y=dframe.target
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
print(x_train.shape)
```

```python
print(x_test.shape)
```

```python
from sklearn.svm import SVC
model=SVC()
model.fit(x_train,y_train)
```

```python
ypre=model.predict(x_test)
ypre
```

```python
model.score(x_test,y_test)
```

```python
modelreg=SVC(C=10)                #it is regular .... means just regular
modelreg.fit(x_train,y_train)
modelreg.score(x_test,y_test)
```

```python
modelreg=SVC(gamma=10)                #it is regular .... means just regular
modelreg.fit(x_train,y_train)
modelreg.score(x_test,y_test)
```

```python
modelreg=SVC(kernel='linear')                #it is regular .... means just regular
modelreg.fit(x_train,y_train)
modelreg.score(x_test,y_test)
```
